## Find the outliers and replace the outliers

```python
from scipy import stats
```

```python
z = np.abs(stats.zscore(data['fare']))
threshold = 3
outliners = data[z>threshold]
```

```python
outliners
```

```python
Q1 = np.percentile(data.fare,25)
Q3 = np.percentile(data.fare,75)

IQR = Q3-Q1
```

```python
IQR
```

```python
#Upper Bound
ub = Q3 + 1.5*IQR
ub_arr = np.array(data.fare>ub) #Boolean values indigating if they are above the upper bound
```

```python
ub_arr.sum()
```

```python
ub
```

```python
#Lower Bound
lb = Q1 - 1.5*IQR
lb_arr = np.array(data.fare<lb) # Boolean valeus indicating they are below the lower bound
```

```python
lb
```

```python
lb_arr.sum()
```

```python
data.fare.shape
```

```python
#Removing the outliners
ub_arr = np.where(data.fare>ub)[0] # Provides us with the indexes
lb_arr = np.where(data.fare<lb)[0] # Provides us with the indexes
data.fare.drop(index = ub_arr,inplace=True)
data.fare.drop(index = lb_arr,inplace=True)
```

```python
data.fare.shape
```

```python
# Replacing the outliners
median = data.loc[data.fare<13,'fare'].median()
data.loc[data.fare>13,'fare'] = np.nan
data.fillna(median,inplace=True)
```
